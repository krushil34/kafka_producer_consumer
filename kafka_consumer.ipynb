```python
topic = 'topic_0'

# consumer subscribed to the topic
consumer.subscribe([topic])

# defining the function to receive the message 
def process_message(message):
  try:
    if message.error():
      if message.error().code() == KafkaError._PARTITION_EOF:
        print(f'End of partition reached {message.topic()} [{message.partition()}] at offset {message.offset()}')
      else:
        raise KafkaException(message.error())
    else:
      key = message.key().decode('utf-8')
      value = json.loads(message.value().decode('utf-8'))

      print(f"Recived message : key {key}, value : {value}")

  except Exception as e:
    print(f'Error consuming/processing message : {e}')
```

```python
# here the code for continuously receive the message
try :
  print(" Listening for messages, press Ctrl + C to exit")

  while True:
    message = consumer.poll(timeout=1)
    if(message):
      process_message(message)

except KeyboardInterrupt:
  print("Interrupted by user, shutting down consumer")
finally:
  consumer.close()
```
