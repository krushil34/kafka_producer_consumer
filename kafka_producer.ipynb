```python
topic = "topic_0"

# loading the json file to send message
with open('customer.json', 'r') as file:
  customer_data = json.load(file)

# defining the callback function
def delivery_status(err,msg):
  if (err):
    print(f"message is not delivered : {err}")
  else:
    print(f"message is delivered to topic {msg.topic()} and partition [{msg.partition()}] at offset {msg.offset()}")

#sending message in a loop 
for record in customer_data:
  try :
    message_value = json.dumps(record)
    message_key = str(int(record['customer_id']+ 11)).encode('utf-8')

    producer.produce(topic, key=message_key, value= message_value, callback = delivery_status)
    producer.poll(1)
  except Exception as e:
    print(f"Error accured : {e}")
producer.flush()

print("all the messages are delivered to kafka")
```
